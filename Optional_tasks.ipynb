**Audio to Text**

```python
!pip -q install openai

from google.colab import files
from getpass import getpass
from openai import OpenAI


api_key = getpass("Enter OpenAI API Key: ")
client = OpenAI(api_key=api_key)

uploaded = files.upload()
audio_file = next(iter(uploaded))

# Speech to text
with open(audio_file, "rb") as f:
    transcript = client.audio.transcriptions.create(
        model="gpt-4o-mini-transcribe",
        file=f
    )

print("Transcribed Text:\n", transcript.text)
```

**Summarize the audio**

```python
!pip -q install openai

from google.colab import files
from getpass import getpass
from openai import OpenAI

api_key = getpass("Enter OpenAI API Key: ")
client = OpenAI(api_key=api_key)


uploaded = files.upload()
audio_file = next(iter(uploaded))

with open(audio_file, "rb") as f:
    transcript = client.audio.transcriptions.create(
        model="gpt-4o-mini-transcribe",
        file=f
    )

text = transcript.text

# Summarize transcript
response = client.responses.create(
    model="gpt-5.4",
    input=f"Summarize the following transcript in 4-5 concise bullet points:\n\n{text}"
)

summary = response.output_text

print("Transcript:\n", text)
print("\nSummary:\n", summary)
```

```python
from deep_translator import GoogleTranslator

def translate_text():
    print("Simple Translator (works like Google Translate)")
    print("Type 'quit' to exit\n")

    while True:
        text = input("Enter text: ").strip()

        if text.lower() in {"quit", "exit"}:
            print("Goodbye!")
            break

        if not text:
            print("Please enter some text.\n")
            continue

        target_lang = input("Enter target language code: ").strip().lower()

        if not target_lang:
            print("Please enter a target language.\n")
            continue

        try:
            translated = GoogleTranslator(source='auto', target=target_lang).translate(text)

            print("\n--- Translation Result ---")
            print("Original:", text)
            print("Translated:", translated)
            print()

        except Exception as e:
            print(f"\nError: {e}\n")


if __name__ == "__main__":
    translate_text()
```
